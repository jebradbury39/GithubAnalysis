```python
def sample_672(dset):
    dset = dset[(dset['type'] == 'PullRequestReviewCommentEvent') | (dset['type'] == 'PullRequestEvent')]
    if (len(dset) > 672):
        dset = dset.sample(n = 672)
    return dset

def update_dataset(day):
    frames = []
    for hour in range (0,24):
        dset = pd.read_json("/share/GithubAnalysis/2016-12-"+ day +"-"+str(hour)+".json", lines = True)
        sampled_dset = sample_672(dset)
        frames.append(sampled_dset)
    full_day_frame = pd.concat(frames)
    full_day_frame.to_csv('/share/GithubAnalysis/pullFiles/2016-12-'+day+'.csv')

for day in range (1,32):
    if day < 10:
        day = '0'+str(day)
    else:
        day = str(day)
    update_dataset(day)
```
